```python
#Exercise 41
from pyspark.sql.functions import sum, avg, count, min, max,col
payments.agg(sum("amount").alias("Total_revenue")).show()
```

```python
#Exercise 42
payments.agg(avg("amount").alias("Avg_amount")).show()
```

```python
#Exercise 43
payments.agg(max("amount").alias("max_amount")).show()
```

```python
#Exercise 44
payments.agg(min("amount").alias("min_amount")).show()
```

```python
#Exercise 45
customers.groupby("city").count().select(col("city"), col("count").alias("customer_count")).show()
```

```python
#Exercise 46
products.groupby("category").agg(count("*").alias("product_count")).show()
```

```python
#Exercise 47
products.groupby("category").agg(avg("price").alias("avg_price")).show()
```

```python
#Exercise 48
order_items.groupby("product_id").agg(sum("quantity").alias("qty_per_prod")).show()
```
