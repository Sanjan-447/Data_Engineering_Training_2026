```python
#Exercise 28
from pyspark.sql.functions import udf
def cust_age(age):
  if age < 28:
    return "Young"
  elif age >= 28 and age <=38:
    return "Adult"
  else:
    return "Senior"

age_group_udf=udf(cust_age)
df_cust_age= customers.withColumn("age_group", age_group_udf(col("age")))
df_cust_age.show()
```
