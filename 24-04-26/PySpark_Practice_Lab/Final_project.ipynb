```python
from pyspark.sql.functions import col ,count, max, min, avg, sum, rank
from pyspark.sql.window import Window
#Exercise 1
final= customers.join(orders, "customer_id")\
         .join(payments, "order_id")\
         .join(order_items, "order_id")\
         .join(products, "product_id" )
final.show()
```

```python
#Exercise 2
final.withColumn("revenue", col("quantity") * col("price"))\
     .show()
```

```python
#Exercise 3
final.withColumn("revenue", col("quantity") * col("price"))\
     .groupby("customer_id", "name")\
     .agg(sum("revenue").alias("total_revenue"))\
     .orderBy(col("total_revenue").desc()).show()
```

```python
#Exercise 4
final.withColumn("revenue", col("quantity") * col("price"))\
     .groupby("product_id", "product_name")\
     .agg(sum("revenue").alias("total_revenue"))\
     .orderBy(col("total_revenue").desc()).show()
```

```python
#Exercise 5
final.withColumn("revenue", col("quantity") * col("price"))\
     .groupby("city")\
     .agg(sum("revenue").alias("total_revenue"))\
     .orderBy(col("total_revenue").desc()).show()
```

```python
#Exercise 6
final_df=customers.join(orders, "customer_id")\
       .join(order_items, "order_id")\
       .join(products, "product_id")\
       .withColumn("revenue", col("quantity") * col("price"))\
       .groupby("customer_id", "name")\
       .agg(sum("revenue").alias("total_revenue"))\
       .orderBy(col("total_revenue").desc())

window= Window.orderBy(col("total_revenue").desc())
final_df.withColumn("cust_rank", rank().over(window)).show()
```

```python
dataset=customers.join(orders, "customer_id")\
         .join(payments, "order_id")\
         .join(order_items, "order_id")\
         .join(products, "product_id" )\
         .withColumn("revenue", col("quantity") * col("price"))

dataset.write.mode("overwrite").option("header", True).csv("ecommerce_report")
print("Saved successfully to eccomerce_report")
```
