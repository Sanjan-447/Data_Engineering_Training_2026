```python
#Exercise 98
spark.sql("""select o.order_id, sum(o.quantity * p.price) as revenue
from products p inner join order_items o
on p.product_id = o.product_id
group by o.order_id
order by revenue desc """).show()
```

```python
#Exercise 99
spark.sql("""
SELECT
    c.customer_id,
    c.name,
    SUM(oi.quantity * p.price) AS total_revenue
FROM customers c
JOIN orders o ON c.customer_id = o.customer_id
JOIN order_items oi ON o.order_id = oi.order_id
JOIN products p ON oi.product_id = p.product_id
GROUP BY c.customer_id, c.name
ORDER BY total_revenue DESC
LIMIT 3
""").show()
```

```python
#Exercise 100
spark.sql("""select  p.product_name, sum(o.quantity * p.price) as revenue
from products p join order_items o
on p.product_id = o.product_id
group by p.product_name
order by revenue desc limit 1 """).show()
```
