```python
#Exercise 71
from pyspark.sql.window import Window
from pyspark.sql.functions import rank, dense_rank, col, row_number

window_spec = Window.orderBy(col("price").desc())
products.withColumn("price_rank", rank().over(window_spec)).show()
```

```python
#Exercise 72
window = Window.partitionBy("category").orderBy(col("price").desc())
products.withColumn("category_rank", rank().over(window)).show()
```

```python
#Exercise 73
window = Window.orderBy(col("signup_date"))
customers.withColumn("signup_row_num", row_number().over(window)).show()
```

```python
#Exercise 74
from pyspark.sql.functions import sum,avg,max,min,count, col
cust_rev= customers.join(orders, on="customer_id", how="inner")\
          .join(order_items, on="order_id", how="inner")\
          .join(products, on="product_id", how="inner")\
          .withColumn("revenue", col("quantity").cast("int") * col("price").cast("int"))\
          .groupby("customer_id", "name")\
          .agg(sum("revenue").alias("total_revenue"))

window = Window.orderBy(col("total_revenue").desc())

cust_rev.withColumn("cust_rank", rank().over(window)).show()
```

```python
#Exercise 75
cust_rev= customers.join(orders, on="customer_id", how="inner")\
          .join(order_items, on="order_id", how="inner")\
          .join(products, on="product_id", how="inner")\
          .withColumn("revenue", col("quantity").cast("int") * col("price").cast("int"))\
          .groupby("customer_id", "name")\
          .agg(sum("revenue").alias("total_revenue"))

window = Window.orderBy(col("total_revenue").desc())

cust_rev.withColumn("cust_dense_rank", dense_rank().over(window)).show()
```

```python
#Exercise 76
cust_rev= customers.join(orders, on="customer_id", how="inner")\
          .join(order_items, on="order_id", how="inner")\
          .join(products, on="product_id", how="inner")\
          .withColumn("revenue", col("quantity").cast("int") * col("price").cast("int"))\
          .groupby("city","customer_id", "name")\
          .agg(sum("revenue").alias("total_revenue"))

window = Window.partitionBy("city").orderBy(col("total_revenue").desc())

cust_rev.withColumn("rank_per_city", dense_rank().over(window))\
        .filter(col("rank_per_city") == 1)\
        .drop("rank_per_city")\
        .show()
```

```python
#Exercise 77
prod_rev= products.join(order_items, on="product_id", how="inner")\
        .withColumn("Revenue", col("quantity") * col("price"))\
        .groupby("category","product_id", "product_name")\
        .agg(sum("Revenue").alias("Total_revenue"))

window = Window.partitionBy("category").orderBy("Total_revenue")

prod_rev.withColumn("rank_per_prod", rank().over(window))\
        .filter(col("rank_per_prod") == 1)\
        .drop("rank_per_prod").show()
```

```python
#Exercise 78
daily_rev = orders.join(order_items, "order_id") \
    .join(products, "product_id") \
    .withColumn("revenue", col("quantity") * col("price")) \
    .groupBy("order_date") \
    .agg(sum("revenue").alias("daily_revenue")) \
    .orderBy("order_date")

# Running total window
window = Window.orderBy("order_date").rowsBetween(Window.unboundedPreceding, Window.currentRow)

daily_rev.withColumn("running_total_revenue", sum("daily_revenue").over(window)) \
         .show()
```

```python
#Exercise 79

daily_rev= orders.join(order_items, "order_id")\
           .join(products, "product_id")\
           .withColumn("revenue", col("quantity") * col("price"))\
           .groupby("product_name", "order_date")\
           .agg(sum("revenue").alias("daily_revenue"))\
           .orderBy("product_name", "order_date")

window= Window.partitionBy("product_name")\
        .orderBy("order_date")\
        .rowsBetween(Window.unboundedPreceding, Window.currentRow)

daily_rev.withColumn("running_total_rev", sum("daily_revenue").over(window))\
         .show()
```
