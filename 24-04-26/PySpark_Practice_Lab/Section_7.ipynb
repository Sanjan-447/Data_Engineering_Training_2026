```python
#Exercise 61
from pyspark.sql.functions import sum, max, min, avg, col, count
orders.groupby("customer_id").agg(count("*").alias("orders_per_cust")).show()
```

```python
#Exercise 62
orders.join(customers, on="customer_id", how="inner")\
      .groupby("city")\
      .agg(count("order_id").alias("order_count"))\
      .show()
```

```python
#Exercise 63
products.join(order_items, on="product_id", how="inner")\
        .withColumn("Revenue", col("quantity") * col("price"))\
        .groupby("product_name")\
        .agg(sum("Revenue").alias("Total_revenue"))\
        .show()
```

```python
#Exercise 64
products.join(order_items, on="product_id", how="inner")\
        .withColumn("Revenue", col("quantity") * col("price"))\
        .groupby("category")\
        .agg(sum("Revenue").alias("Total_revenue"))\
        .show()
```

```python
#Exercise 65
customers.join(orders, on="customer_id", how="inner")\
         .join(order_items, on="order_id", how="inner")\
         .join(products, on="product_id", how="inner")\
         .withColumn("revenue", col("quantity") * col("price"))\
         .groupby("city")\
         .agg(sum("revenue").alias("total_revenue"))\
         .show()
```

```python
#Exercise 66
products.join(order_items, on='product_id', how="inner")\
        .groupby("product_id", "product_name")\
        .agg(sum("quantity").alias("Total_qty"))\
        .show()
```

```python
#Exercise 67
products.join(order_items, on='product_id', how="inner")\
        .groupby("category")\
        .agg(sum("quantity").alias("Total_qty"))\
        .show()
```

```python
#Exercise 68
customers.join(orders, on="customer_id", how="inner")\
         .join(order_items, on="order_id", how="inner")\
         .join(products, on="product_id", how="inner")\
         .withColumn("revenue", col("quantity") * col("price"))\
         .groupby("name","customer_id")\
         .agg(sum("revenue").alias("total_revenue"))\
         .show()
```

```python
#Exercise 69
customers.join(orders, on="customer_id", how="inner")\
         .join(order_items, on="order_id", how="inner")\
         .join(products, on="product_id", how="inner")\
         .withColumn("revenue", col("quantity") * col("price"))\
         .groupby("name","customer_id")\
         .agg(sum("revenue").alias("total_revenue"))\
         .orderBy(col("total_revenue").desc())\
         .show(5)
```

```python
#Exercise 70
products.join(order_items, on="product_id", how="inner")\
        .withColumn("Revenue", col("quantity") * col("price"))\
        .groupby("product_id", "product_name")\
        .agg(sum("Revenue").alias("Total_revenue"))\
        .orderBy(col("Total_revenue").desc())\
        .show(3)
```
