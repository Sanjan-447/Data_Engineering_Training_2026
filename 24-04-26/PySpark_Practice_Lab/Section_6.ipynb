```python
#Exercise 52
from pyspark.sql.functions import col, sum, avg, max, min, count
customers.join(orders, on="customer_id", how="inner").select(col("name").alias("customer_name"), col("status").alias("order_status")).show()
```

```python
#Exercise 53
order_items.join(orders, on="order_id", how="inner").show()
```

```python
#Exercise 54
products.join(order_items, on="product_id", how="inner").show()
```

```python
#Exercise 55
products.join(order_items, on="product_id", how="inner").withColumn("revenue", col("quantity") * col("price")).show()
```

```python
#Exercise 56
customers.join(orders, on="customer_id", how="inner").join(order_items, on="order_id", how="inner").join(products, on="product_id", how="inner").show()
```

```python
#Exercise 57
customers.join(orders, on="customer_id", how="inner")\
         .join(order_items, on="order_id", how="inner")\
         .join(products, on="product_id", how="inner")\
         .select(col("name").alias("customer_name"), col("product_name"),col("quantity"))\
         .show()
```

```python
#Exercise 58
products.join(order_items, on="product_id", how="inner")\
        .withColumn("Total_revenue",col("quantity") * col("price"))\
        .groupby("order_id")\
        .agg(sum("Total_revenue").alias("rev_per_order"))\
        .show()
```

```python
#Exercise 59
products.join(order_items, on="product_id", how="inner")\
        .withColumn("Revenue", col("quantity") * col("price"))\
        .groupby("product_name")\
        .agg(sum("Revenue").alias("Total_revenue"))\
        .show()
```

```python
#Exercise 60
customers.join(orders, on="customer_id", how="inner")\
         .join(order_items, on="order_id", how="inner")\
         .join(products, on="product_id", how="inner")\
         .withColumn("revenue", col("quantity") * col("price"))\
         .groupby("customer_id","name")\
         .agg(sum("revenue").alias("Total_revenue"))\
         .show()
```
