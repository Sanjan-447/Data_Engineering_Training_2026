```python
#Exercise 61
final_df = visits_df.join(patients_df, "patient_id") \
                    .join(doctors_df, "doctor_id") \
                    .join(payments_df, "visit_id")
final_df.createOrReplaceTempView("final_visits")
spark.sql("""
create or replace table final_visits_delta
using delta as
select * from final_visits
""")
spark.sql("select * from final_visits_delta").show()
```

```python
#Exercise 66
spark.sql("""
create or replace table clean_completed_visits
using delta as
select *
from final_visits_delta
where visit_status = 'Completed'
  and payment_status = 'Paid'
""")
```
