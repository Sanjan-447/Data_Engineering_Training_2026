```python
import dlt
@dlt.table(
    name= "bronze_patients"
)

def bronze_patients():
  return spark.createDataFrame(patients_data, patients_columns)

@dlt.table(
    name="bronze_doctors"
)

def bronze_doctors():
  return spark.createDataFrame(doctors_data, doctors_columns)

@dlt.table(
    name="bronze_visits"
)

def bronze_visits():
  return spark.createDataFrame(visits_data, visits_columns)

@dlt.table(
    name="bronze_payments"
)

def bronze_payments():
  return spark.createDataFrame(payments_data, payments_columns)
```

```python
@dlt.table(name="silver_patient_visits")
def silver_patient_visits():

    patients = dlt.read("bronze_patients")
    doctors = dlt.read("bronze_doctors")
    visits = dlt.read("bronze_visits")
    payments = dlt.read("bronze_payments")

    df = visits \
        .join(patients, "patient_id", "left") \
        .join(doctors, "doctor_id", "left") \
        .join(payments, "visit_id", "left")

    df = df.filter(col("visit_status").isNotNull()) \
           .filter(col("bill_amount") > 0)

    df = df.withColumn("total_bill", col("bill_amount") + col("tests_count") * 2000)

    return df
```

```python
@dlt.table(name="gold_city_revenue")
def gold_city_revenue():
    df = dlt.read("silver_patient_visits")

    return df.groupBy("city") \
             .agg(sum("total_bill").alias("city_revenue"))
```

```python
@dlt.table(name="gold_specialization_revenue")
def gold_specialization_revenue():
    df = dlt.read("silver_patient_visits")

    return df.groupBy("specialization") \
             .agg(sum("total_bill").alias("specialization_revenue"))
```
