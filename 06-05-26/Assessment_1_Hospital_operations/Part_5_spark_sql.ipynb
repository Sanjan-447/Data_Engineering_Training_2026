```python
#Exercise 43
spark.sql("""
select v.*
from visits v
join doctors d
on v.doctor_id = d.doctor_id
where d.specialization = 'Cardiology'
""").show()
```

```python
#Exercise 44
spark.sql("""
select p.city,sum(pay.bill_amount) as total_revenue
from visits v
join patients p on v.patient_id = p.patient_id
join payments pay on v.visit_id = pay.visit_id
group by p.city
order by total_revenue DESC
""").show()
```

```python
#Exercise 45
spark.sql("""
select d.specialization, sum(pay.bill_amount) as total_revenue
from visits v
join doctors d on v.doctor_id = d.doctor_id
join payments pay on v.visit_id = pay.visit_id
group by d.specialization
order by total_revenue DESC
""").show()
```

```python
#Exercise 47
spark.sql("""
select d.doctor_name, count(v.visit_id) as total_visits
from visits v
join doctors d on v.doctor_id = d.doctor_id
group by d.doctor_name
order by total_visits desc
""").show()
```

```python
#Exercise 48
spark.sql("""
SELECT p.patient_name,
       COUNT(v.visit_id) AS total_visits
FROM visits v
JOIN patients p ON v.patient_id = p.patient_id
GROUP BY p.patient_name
ORDER BY total_visits DESC
""").show()
```

```python
#Exercise 49
spark.sql("""
SELECT payment_mode,
       AVG(bill_amount) AS avg_bill
FROM payments
GROUP BY payment_mode
""").show()
```

```python
#Exercise 50
spark.sql("""
select p.patient_name,sum(pay.bill_amount) as total_billing
from visits v join patients p on v.patient_id = p.patient_id
join payments pay on v.visit_id = pay.visit_id
group by p.patient_name
having sum(pay.bill_amount) > 10000
order by total_billing desc
""").show()
```
