```python
import delta
from pyspark.sql import SparkSession
from delta import configure_spark_with_delta_pip

builder = SparkSession.builder \
    .appName("Part_9_parquet") \
    .config("spark.sql.extensions", "io.delta.sql.DeltaSparkSessionExtension") \
    .config("spark.sql.catalog.spark_catalog", "org.apache.spark.sql.delta.catalog.DeltaCatalog")

spark = configure_spark_with_delta_pip(builder).getOrCreate()
print(delta.__version__)
print("Spark session ready")
```

```python
#Exercise 81
patients_df.write.mode("overwrite").parquet("/tmp/patients_parquet")
```

```python
#Exercise 82
parquet_df= spark.read.parquet("/tmp/patients_parquet")
parquet_df.show()
```

```python
#Exercise 83
parquet_df.write.format("delta").mode("overwrite").save("/tmp/patients_delta")
```

```python
#Exercise 84
delta_df = spark.read.format("delta").load("/tmp/patients_delta")
delta_df.show()
```

```python
#Exercise 85
print("PARQUET FILES:")
print(parquet_df.count())

print("DELTA FILES:")
print(delta_df.count())
```

```python
#Exercise 86
from delta.tables import DeltaTable

delta_table = DeltaTable.forPath(spark, "/tmp/patients_delta")

delta_table.update(
    condition="membership = 'Bronze'",
    set={"membership": "'Silver'"}
)
```
