```python
#Exercise 14
visits_df= visits_df.join(doctors_df.select("doctor_id", "consultation_fee"), "doctor_id") \
                     .withColumn("estimated_total_bill", col("consultation_fee") + col("test_cost"))
visits_df.show()
```

```python
#Exercise 15
def age_category(age):
    if age < 30:
        return "Young"
    elif age <= 45:
        return "Adult"
    else:
        return "Senior"

age_category_udf = udf(age_category)

patients_df.withColumn("age_category", age_category_udf(col("age"))).show()
```

```python
#Exercise 16
def membership_priority(mem):
    if mem == "Gold":
        return "High"
    elif mem == "Silver":
        return "Medium"
    else:
        return "Low"

membership_priority_udf = udf(membership_priority)

patients_df.withColumn("membership_priority", membership_priority_udf(col("membership"))).show()
```

```python
#Exercise 17
def visit_priority(tests):
    if tests >= 3:
        return "High"
    elif tests == 2:
        return "Medium"
    else:
        return "Low"

visit_priority_udf = udf(visit_priority)

visits_df.withColumn("visit_priority", visit_priority_udf(col("tests_count"))).show()
```

```python
#Exercise 18
def high_bill_flag(bill):
    return bill > 8000

high_bill_flag_udf = udf(high_bill_flag)

visits_df.withColumn("high_bill_flag", high_bill_flag_udf(col("estimated_total_bill"))).show()
```
