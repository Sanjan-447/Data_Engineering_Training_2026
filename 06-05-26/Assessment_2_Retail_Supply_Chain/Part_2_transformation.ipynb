```python
#Exercise 11
from pyspark.sql.functions import col, sum, count, max, min, avg, when, upper, udf, to_date

orders_df.withColumn("order_date", to_date(col("order_date"))).show()
```

```python
#Exercise 12
orders_df.join(products_df, "product_id")\
         .withColumn("total_order_value", col("price") * col("quantity"))\
         .select("product_name", "price", "quantity", "total_order_value")\
         .show()
```

```python
#Exercise 13
def stock_status(stock_quantity):
  if stock_quantity >= 100:
    return "High stock"
  elif stock_quantity >= 50:
    return "Medium stock"
  else:
    return "Low stock"

stock_status_udf = udf(stock_status)

products_df.withColumn("stock_status", stock_status_udf(col("stock_quantity"))).show()
```

```python
#Exercise 14
def priority(quantity):
  if quantity >= 50:
    return "High"
  elif quantity >= 20:
    return "Medium"
  else:
    return "Low"

priority_udf = udf(priority)
orders_df.withColumn("order_priority", priority_udf(col("quantity"))).show()
```

```python
#Exercise 15
def expensive(price):
  if price >= 25000:
    return "Expensive"
  elif price >=8000 and price <= 24999:
    return "Moderate"
  else:
    return "Affordable"
exp_udf= udf(expensive)
products_df.withColumn("expensive_product_flag", exp_udf(col("price"))).show()
```
