```python
#Exercise 96
create catalog if not exists orders_catalog;
```

```python
#Exercise 97
create schema if not exists orders_catalog.orders_schema;
```

```python
#Exercise 98
use orders_catalog.orders_schema;
create table if not exists orders_catalog.orders_schema.final_delta using delta location '/tmp/final_delta';
```

```python
#Exercise 99
CREATE OR REPLACE TABLE orders_catalog.sales_schema.city_revenue AS
SELECT warehouse_city,
       SUM(bill_amount) AS total_revenue
FROM orders_catalog.orders_schema.final_delta
GROUP BY warehouse_city;
```

```python
#Exercise 100
grant select on table orders_catalog.orders_schema.city_revenue to 'users';

grant usage on schema orders_catalog.orders_schema to 'users';

grant usage on catalog orders_catalog to 'users';
```

Unity Catalog ensures:

Only authorized users can access catalog/schema/table.
Permissions are centrally controlled.
Lineage is tracked automatically.
Data is secured across workspaces using one governance layer.
