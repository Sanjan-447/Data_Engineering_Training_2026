```python
from pyspark.sql import SparkSession
spark = SparkSession.builder\
        .appName("Part_1_Dataframe")\
        .getOrCreate()
print("Spark Session ready")
```

```python
data = [
(101,"Arjun Reddy","Hyderabad","Cardiology",5000,1),
(102,"Sneha Kapoor","Delhi","Orthopedics",3000,2),
(103,"Rahul Sharma","Mumbai","Dermatology",1500,1),
(104,"Priya Nair","Bangalore","Cardiology",5000,2),
(105,"Vikram Singh","Chennai","Neurology",7000,1),
(106,"Ananya Das","Kolkata","Orthopedics",3000,3),
(107,"Karan Patel","Ahmedabad","Cardiology",5000,1),
(108,"Meera Iyer","Bangalore","Dermatology",1500,2)
]
columns = [
"visit_id",
"patient_name",
"city",
"department",
"consultation_fee",
"tests_count"
]

df = spark.createDataFrame(data, columns)
df.show()
```

```python
from pyspark.sql.functions import col, count, sum, max, min, avg
df_cost= df.withColumn("test_cost", col("tests_count") * 2000)\
  .withColumn("total_cost", col("consultation_fee") + col("test_cost"))

df_cost.show()
```

```python
df_cost.filter(col("total_cost") > 8000).show()
```

```python
df.groupby("department").agg(count("*").alias("visit_count")).show()

df_cost.groupby("department").agg(sum("total_cost").alias("total_revenue")).show()

df_cost.groupby("department").agg(sum("tests_count").alias("count_per_dept")).show()
```

```python
df_cost.orderBy("total_cost").show()

df_cost.orderBy(col("total_cost").desc()).show()

```
