```python
from pyspark.sql import SparkSession
spark = SparkSession.builder\
        .appName("Part_2_spark_sql")\
        .getOrCreate()
print("Spark Session ready")
```

```python
data = [
(101,"Arjun Reddy","Hyderabad","Cardiology",5000,1),
(102,"Sneha Kapoor","Delhi","Orthopedics",3000,2),
(103,"Rahul Sharma","Mumbai","Dermatology",1500,1),
(104,"Priya Nair","Bangalore","Cardiology",5000,2),
(105,"Vikram Singh","Chennai","Neurology",7000,1),
(106,"Ananya Das","Kolkata","Orthopedics",3000,3),
(107,"Karan Patel","Ahmedabad","Cardiology",5000,1),
(108,"Meera Iyer","Bangalore","Dermatology",1500,2)
]
columns = [
"visit_id",
"patient_name",
"city",
"department",
"consultation_fee",
"tests_count"
]

df = spark.createDataFrame(data, columns)
df.show()
```

```python
df.createOrReplaceTempView("patient_visits")
```

```python
spark.sql(" select * from patient_visits").show()
```

```python
spark.sql("""select * from patient_visits where department="Cardiology" """).show()
```

```python
spark.sql(""" select city, sum(consultation_fee + (tests_count * 2000)) as total_revenue from patient_visits group by city """).show()
```

```python
spark.sql("""
SELECT patient_name, city, department,
       (consultation_fee + (tests_count * 2000)) AS total_bill
FROM patient_visits
ORDER BY total_bill DESC
LIMIT 3
""").show()
```

```python
spark.sql(""" select department, count(*) as Patient_count from patient_visits group by department""").show()
```
