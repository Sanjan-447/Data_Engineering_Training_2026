```sql
select * from employees order by salary desc limit 3;
```

```sql
select department, avg(salary) as avg_salary from employees group by department order by avg_salary desc limit 1;
```

```sql
select city, count(*) as emp_count from employees group by city order by emp_count desc limit 1;
```

```sql
select * from employees e where salary > (select avg(salary) from employees where department = e.department);
```

```sql
select * from (select * , rank() over (partition by department order by experience desc ) as rank from employees) where rank =1;
```
