```sql
USE training_sql_lab;
```

```sql
create table employee_salary_analysis as select emp_name, city, department, salary, experience from employees;
```

```sql
alter table employee_salary_analysis add column salary_category varchar(50);
alter table employee_salary_analysis add column department_avg_salary double;
```

```sql
update employee_salary_analysis set salary_category = 
case  when salary >= 95000 then "High"
      when salary >= 80000 then "Medium"
      else "Low"
end; 
```

```sql
update employee_salary_analysis e set department_avg_salary = 
(select avg(salary) from employee_salary_analysis where department = e.department);
```

```sql
select * from employee_salary_analysis;
```
