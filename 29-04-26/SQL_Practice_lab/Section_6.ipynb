```sql
select e.*, p.budget from employees e join projects p on e.department = p.department 
```

```sql
select e.emp_name, e.department, p.project_name from employees e join projects p on e.department = p.department 
```

```sql
select department, sum(budget) as total_budget from projects group by department;
```

```sql
select distinct e.* from employees e join projects p on e.department= p.department;
```

```sql
select p.project_name, count(*) as emp_count from employees e join projects p on e.department=p.department GROUP BY p.project_name;
```
