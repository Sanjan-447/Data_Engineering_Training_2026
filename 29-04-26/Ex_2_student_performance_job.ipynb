```python
from pyspark.sql.functions import col, avg, when, udf, count
data = [
(101, "Arjun Kumar", "Python", 86, 92),
(102, "Neha Gupta", "Python", 74, 80),
(103, "Rahul Nair", "Data Engineering", 88, 85),
(104, "Priya Sharma", "AI", 91, 95),
(105, "Siddharth Rao", "Data Engineering", 67, 72),
(106, "Ayesha Khan", "AI", 79, 88),
(107, "Kavya Reddy", "Python", 94, 96),
(108, "Manish Yadav", "Data Engineering", 58, 69),
(109, "Divya Menon", "AI", 83, 90),
(110, "Farhan Ali", "Python", 62, 75)
]
columns = ["student_id", "student_name", "course", "marks", "attendance"]
df = spark.createDataFrame(data, columns)
display(df)
```

```python
def mark_grade(marks):
    if marks >= 90:
        return "A"
    elif  marks >=75:
        return "B"
    elif marks >= 60:
        return "C"
    else:
        return "Needs improvement"
mark_udf= udf(mark_grade)
df.withColumn("grade", mark_udf(col("marks"))).show()
```

```python
def cert_status(marks, attendance):
    if marks >= 75 and attendance >= 80:
        return "Eligible"
    else:
        return "Not Eligible"
cert_udf = udf(cert_status)
cert_df=df.withColumn("certification_status", cert_udf(col("marks"), col("attendance")))
cert_df.show()
    
```
