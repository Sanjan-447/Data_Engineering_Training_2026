```python
#Exercise 61
from pyspark.sql.functions import count, col, sum, avg, max, min, lit
clean_df=df.dropDuplicates()
clean_df.show()
```

```python
#Exercise 62
clean_df.select(
    [sum(col(c).isNull().cast("int")).alias(c + "_nulls") for c in df.columns]
).show()
```

```python
#Exercise 63
clean_df=clean_df.fillna({"salary": 0})
clean_df.show()
```

```python
#Exercise 64
clean_df=clean_df.na.drop(subset=["department"])
clean_df.show()
```

```python
#Exercise 65
avg_exp= df.select(avg("experience")).collect()[0][0]
clean_df=clean_df.fillna({"department": avg_exp})
clean_df.show()
```

```python
#Exercise 67
clean_df=clean_df.fillna({"department": "Unknown"})
clean_df.show()
```
