```python
df["timestamp"] = df["timestamp"].astype(str).str.strip()

df["timestamp"] = pd.to_datetime(df["timestamp"], errors="coerce", dayfirst=True)

df = df.dropna(subset=["timestamp"])

df["energy_kwh"] = df["energy_kwh"].astype(float)

print("cleaned dataset: ")
print(df)
```

```python
df["date"] = df["timestamp"].dt.date
df["month"] = df["timestamp"].dt.to_period("M")
print(df.head())
```

```python
daily_total = df.groupby("date")["energy_kwh"].sum()

monthly_total = df.groupby("month")["energy_kwh"].sum()

daily_total_np = np.array(daily_total)
monthly_total_np = np.array(monthly_total)

daily_avg_np = np.mean(daily_total_np)
monthly_avg_np = np.mean(monthly_total_np)

print("\nDaily Energy Total (kWh):")
print(daily_total)

print("\nMonthly Energy Total (kWh):")
print(monthly_total)

print("\nAverage Daily Energy Usage (NumPy):", daily_avg_np)
print("Average Monthly Energy Usage (NumPy):", monthly_avg_np)
```
