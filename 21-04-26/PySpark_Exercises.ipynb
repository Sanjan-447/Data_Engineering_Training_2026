```python
from pyspark.sql import SparkSession

spark = SparkSession.builder\
        .appName("PySpark Exercise")\
        .getOrCreate()

print("Spark Session created.")
```

```python
### Exercise 1
data = [
    ("Amit Sharma", 75000, "IT"),
    ("Priya Verma", 68000, "HR"),
    ("Rahul Mehta", 82000, "Finance"),
    ("Sneha Iyer", 90000, "Marketing"),
    ("Vikram Singh", 72000, "Operations")
]

columns = ["Name", "Salary", "Department"]

df = spark.createDataFrame(data, columns)
df.show()
```

```python
### Exercise 2
df.select("Name", "Salary").show()
```

```python
### Exercise 3
df.filter(df.Salary > 50000).show()
```

```python
### Exercise 4
df.groupby("Department").count().show()
```

```python
### Exercise 5
df.orderBy(df.Salary.desc()).show()
```

```python
df.orderBy(df.Salary).show()
```
