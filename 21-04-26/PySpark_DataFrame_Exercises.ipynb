```python
#PART 6 — Aggregations
#Exercise 21

from pyspark.sql.functions import sum, avg, max, min

df.select(sum("salary").alias("total_salary")).show()
```

```python
#Exercise 22

df.select(avg("salary").alias("avg_salary")).show()
```

```python
#Exercise 23

df.select(max("salary").alias("highest_salary")).show()
```

```python
#Exercise 24

df.select(min("salary").alias("lowest_salary")).show()
```

```python
#PART 7 — GroupBy
#Exercise 25

df.groupBy("department").count().show()
```

```python
#Exercise 26

df.groupBy("department").agg(avg("salary").alias("avg_salary")).show()
```

```python
#Exercise 27

df.groupBy("department").agg(sum("salary").alias("total_salary")).show()
```

```python
#PART 8 — Joining DataFrames
#Exercise 28

df.join(df_dept, on="department", how="inner").show()
```

```python
#Exercise 29

df.join(df_dept, on="department", how="left").show()
```

```python
#Exercise 30

df.join(df_dept, on="department", how="inner").select("name", "department", "location").show()
```

```python
#PART 9 — Updating Data
#Exercise 31

df.withColumn("bonus", df.salary * 0.10).show()
```

```python
#Exercise 32
from pyspark.sql.functions import lit

df.withColumn("Company", lit("BotCampus")).show()
```

```python
#PART 10 — Handling Null Values
#Exercise 33

df.filter(col("department").isNull()).show()
```

```python
#Exercise 34

df.na.fill({"department" : "Unknown"}).show()
```

```python
#Exercise 35

df.na.drop().show()
```

```python
#PART 11 — Built-in Functions
#Exercise 36

from pyspark.sql.functions import upper, lower, length

df.select("name",upper("name").alias("upper_name")).show()
```

```python
#Exercise 37

df.select("name",length("name").alias("length_name")).show()
```

```python
#Exercise 38

df.withColumn("salary_in_lakhs", df.salary/100000).show()
```

```python
#PART 12 — When / Otherwise
#Exercise 39
from pyspark.sql.functions import col, when

df.withColumn("salary_grade", when(col("salary") > 75000, "High").when((col("salary") >= 60000) & (col("salary") <= 74999), "Medium").otherwise("Low")).show()
```

```python
#PART 13 — Date Functions
#Exercise 40

from pyspark.sql.functions import *

df.withColumn("joining_date", to_date("joining_date", "dd-MMM-yyyy")).show()
```
