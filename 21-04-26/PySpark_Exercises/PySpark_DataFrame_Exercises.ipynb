```python
#PART 12 — When / Otherwise
#Exercise 39
from pyspark.sql.functions import col, when, udf

# df.withColumn("salary_grade", when(col("salary") > 75000, "High").when((col("salary") >= 60000) & (col("salary") <= 74999), "Medium").otherwise("Low")).show()

def salary_grade(salary):
  if salary > 75000:
    return "High"
  elif salary >= 60000 and salary <= 74999:
    return "Medium"
  else:
    return "Low"

salary_udf=udf(salary_grade)

df.withColumn("salary_grade", salary_udf(col("salary"))).show()
```

```python
#PART 15 — Window Functions
#Exercise 44
from pyspark.sql.window import Window
from pyspark.sql.functions import rank,col

window = Window.partitionBy("department").orderBy(col("salary").desc())

df_ranked = df.withColumn("salary_rank", rank().over(window))

df_ranked.show()
```

```python
#PART 20 — Widgets in Colab
#Exercise 56
import ipywidgets as widgets
from IPython.display import display

departments = [row["department"] for row in df.select("department").distinct().collect()]

dept_widget = widgets.Dropdown(
    options=departments,
    value=departments[0],
    description="Department :"
)

display(dept_widget)
```

```python
#Exercise 57
salary= [row["salary"] for row in df.select("salary").orderBy(col("salary")).collect()]
salary_widget= widgets. IntSlider(
    value=salary[0],
    min=salary[0],
    max=salary[-1],
    step=1000,
    description= "Salary: "
)
display(salary_widget)
```

```python
#Exercise 58
departments = [row["department"] for row in df.select("department").distinct().collect()]

dept_widget = widgets.Dropdown(
    options=departments,
    value=departments[0],
    description="Department:"
)

def show_employees(change):
    df.filter(df.department == dept_widget.value).show()

dept_widget.observe(show_employees, names="value")

display(dept_widget)

# Show initial result
df.filter(df.department == dept_widget.value).show()
```
