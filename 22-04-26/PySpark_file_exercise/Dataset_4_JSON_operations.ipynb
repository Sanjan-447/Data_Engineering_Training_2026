```python
from pyspark.sql import SparkSession
spark = SparkSession.builder\
        .appName("Dataset_4_JSON_operations")\
        .getOrCreate()
print("Spark Session created")
```

```python
#Question 1
from pyspark.sql.functions import explode, col

df= spark.read.option("multiline", "true").json("orders.json")

df_orders= df.select(explode(col("orders")).alias("order")).select("order.*")
df_orders.show()
```

```python
#Question 2
df_ord=df_orders.select("order_id", "customer", "product", "amount", "city")
df_ord.show()
```

```python
#Question 3
df_ord.count()
```

```python
#Question 4
from pyspark.sql.functions import  sum, avg, min, max
df_ord.select(sum("amount").alias("Total_amount")).show()
```

```python
#Question 5
df_ord.groupby("customer").agg(sum("amount").alias("Total_spent")).show()
```

```python
#Question 6
df_ord.groupby("customer").agg(sum("amount").alias("Total_spent")).orderBy(col("Total_spent").desc()).select(col("customer").alias("Highest_spending_cust")).show(1)
```

```python
#Question 7
df_ord.groupby("product").agg(sum("amount").alias("Total_sales")).show()
```

```python
#Question 8
df_ord.filter(col("city")=="Hyderabad").show()
```

```python
#Question 9
df_ord.filter(col("amount") > 10000).show()
```

```python
#Question 10
df_ord.groupby("city").count().select(col("city"), col("count").alias("Orders_per_city")).show()
```
