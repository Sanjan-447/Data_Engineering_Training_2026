```python
#Question 1

df = spark.read.text("logins.txt")
df.show()
```

```python
df_names = df.withColumnRenamed("value", "Name")
df_names.show()
```

```python
#Question 2

df_names.count()
```

```python
#Question 3
df_names.distinct().show()
```

```python
#Question 4

df_names.groupby("Name").count().show()
```

```python
#Question 5
from pyspark.sql.functions import col, count
df_names.groupBy("Name")\
        .count()\
        .orderBy(col("count").desc())\
        .select("Name")\
        .show(3)
```

```python
#Question 6
df_names.groupby("Name").count().filter(col("count") > 4).show()
```

```python
#Question 7
df_dict= dict(df_names.groupby("Name").count().collect())
print(df_dict)
```
