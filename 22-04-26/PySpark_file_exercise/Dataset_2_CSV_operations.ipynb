```python
#Question 1
df = spark.read.csv("employees.csv", header=True, inferSchema=True)
df.show()
```

```python
#Question 2
df.count()
```

```python
#Question 3
df.filter(df.department == "IT").show()
```

```python
#Question 4
df.filter(df.salary > 75000).show()
```

```python
#Question 5
from pyspark.sql.functions import sum, avg, max, min
df.select(avg("Salary").alias("Average_Salary")).show()
```

```python
#Question 6
df.select(max("Salary").alias("Highest_paid_employee")).show()
```

```python
#Question 7
df.select(min("Salary").alias("Lowest_paid_employee")).show()
```

```python
#Question 8
from pyspark.sql.functions import col
df.groupby("department").count().select("department", col("count").alias("emp_per_dept")).show()
```

```python
#Question 9
df.groupby("department").agg(avg("Salary").alias("Avg_sal_dept")).show()
```

```python
#Question 10
df.groupby("City").count().select("City", col("count").alias("emp_per_city")).show()
```

```python
#Question 11
df.orderBy(col("Salary").desc()).show(5)
```

```python
#Question 11
df.filter((df.city == "Hyderabad") & (df.salary >70000)).show()
```
