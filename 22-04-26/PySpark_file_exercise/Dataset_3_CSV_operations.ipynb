```python
df = spark.read.csv("sales.csv", header=True, inferSchema=True)
df.show()
```

```python
#Question 1
from pyspark.sql.functions import col
df_rev = df.withColumn("revenue", col("quantity")* col("price"))
df_rev.show()
```

```python
#Question 2
from pyspark.sql.functions import sum,max, min, avg

df_rev.select(sum("revenue").alias("Total_Revenue")).show()
```

```python
#Question 3
df_rev.groupBy("product").agg(sum("revenue").alias("Rev_per_product")).show()
```

```python
#Question 4
df_rev.groupBy("product").agg(sum("quantity").alias("qty_per_prod")).show()
```

```python
#Question 5
df_rev.groupBy("product").agg(sum("quantity").alias("Total_qty")).orderBy(col("Total_qty").desc()).show(1)
```

```python
#Question 6
df_rev.groupBy("emp_id").agg(sum("revenue").alias("Total_rev")).orderBy(col("Total_rev").desc()).show(1)
```

```python
#Question 7
df_rev.select(avg("revenue").alias("Avg_sale_value")).show()
```

```python
#Question 8
df_rev.groupby("product").agg(sum("revenue").alias("Total_rev")).filter(col("Total_rev") > 100000).show()
```
