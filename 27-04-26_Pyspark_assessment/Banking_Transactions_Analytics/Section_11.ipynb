```python
from pyspark.sql import SparkSession

spark= SparkSession.builder\
      .appName("Section_11")\
      .getOrCreate()

print("Spark session ready")
```

```python
customer_profiles_json = """[
  {
    "customer_id": 1,
    "name": "Rahul",
    "contact": {"email": "rahul@mail.com", "phone": "9000011111"},
    "services": ["UPI", "Credit Card", "Net Banking"]
  },
  {
    "customer_id": 2,
    "name": "Sneha",
    "contact": {"email": "sneha@mail.com", "phone": "9000022222"},
    "services": ["UPI", "Debit Card"]
  },
  {
    "customer_id": 3,
    "name": "Arjun",
    "contact": {"email": "arjun@mail.com", "phone": "9000033333"},
    "services": ["Net Banking", "Loan"]
  },
  {
    "customer_id": 6,
    "name": "Meera",
    "contact": {"email": "meera@mail.com", "phone": "9000066666"},
    "services": ["UPI", "Credit Card", "Loan"]
  },
  {
    "customer_id": 10,
    "name": "Vikram",
    "contact": {"email": "vikram@mail.com", "phone": "9000101010"},
    "services": ["Net Banking", "Wealth"]
  }
]
"""

with open("customer_profiles.json", "w") as f:
    f.write(customer_profiles_json)

profiles = spark.read.json("customer_profiles.json", multiLine=True)
profiles.show()
```

```python
from pyspark.sql.functions import explode, col, count
profiles.show()
```

```python
profiles.select(col("customer_id"), col("name"), col("contact.email").alias("Email"), col("contact.phone").alias("Phone")).show()
```

```python
profiles.select(explode(col("services")).alias("Service")).show()
```

```python
profiles.select("customer_id","name", explode(col("services")).alias("service")).filter(col("service") == "UPI").show()
```

```python
profiles.select("customer_id","name", explode(col("services")).alias("service"))\
        .groupby(col("service"))\
        .agg(count("*").alias("cust_count"))\
        .show()
```
