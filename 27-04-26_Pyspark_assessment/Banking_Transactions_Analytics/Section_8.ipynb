```python
from pyspark.sql.functions import udf, col

def balance_category(balance):
    if balance >= 100000:
        return "High"
    elif balance >= 50000:
        return "Medium"
    else:
        return "Low"

bal_udf = udf(balance_category)

accounts.withColumn("balance_category", bal_udf(col("balance"))).show()
```

```python
def age_group(age):
    if age < 30:
        return "Young"
    elif age < 40:
        return "Adult"
    else:
        return "Senior"

age_udf = udf(age_group)

customers.withColumn("age_group", age_udf(col("age"))).show()
```

```python
def age_group(age):
    if age < 30:
        return "Young"
    elif age < 40:
        return "Adult"
    else:
        return "Senior"

age_udf = udf(age_group)

customers.withColumn("age_group", age_udf(col("age"))).show()
```

```python
def region_priority(region):
    if region == "South":
        return "High Priority"
    elif region == "North":
        return "Medium Priority"
    elif region == "West":
        return "Watch"
    else:
        return "Unknown"

region_udf = udf(region_priority)

branches.withColumn("region_priority", region_udf(col("region"))).show()
```

```python
def acc_type_label(acc_type):
    if acc_type == "Savings":
        return "Retail"
    elif acc_type == "Current":
        return "Business"
    else:
        return "Unknown"

acc_udf = udf(acc_type_label)

customers.withColumn("account_type_label", acc_udf(col("account_type"))).show()
```
