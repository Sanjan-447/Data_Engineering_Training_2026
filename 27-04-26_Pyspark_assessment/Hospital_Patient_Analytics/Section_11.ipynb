```python
from pyspark.sql import SparkSession

spark= SparkSession.builder\
       .appName("Section_11")\
       .getOrCreate()

sc= spark.sparkContext
print("Spark session created")
```

```python
hospital_logs_txt = """Aarav login
Priya login
Rahul appointment
Sneha login
Aarav payment
Kiran appointment
Meera login
Vikram appointment
Divya payment
Farhan login
Simran appointment
Neha payment
Amit login
Rahul payment
Meera appointment
Aarav logout
Priya payment
Divya login
Vikram payment
Farhan appointment
"""

with open("hospital_logs.txt", "w") as f:
  f.write(hospital_logs_txt)

logs = sc.textFile("hospital_logs.txt")
logs.collect()
```
