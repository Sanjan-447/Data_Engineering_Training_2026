```python
#Exercise 71
from pyspark.sql.functions import col, sum, min, avg, max, min,when, lit, concat, udf

def age_category(age):
  if age < 30:
    return "young"
  elif age >= 30 and age<=45:
    return "Adult"
  else:
    return "Senior"

age_udf= udf(age_category)
patients.withColumn("age_category", age_udf(col("age"))).show()
```

```python
#Exercise 72
def bill_cat(bill_amount):
  if bill_amount < 1000:
    return "Low"
  elif bill_amount >=1000 and bill_amount <= 1600:
    return "Medium"
  else:
    return "High"

bill_udf = udf(bill_cat)
bills.withColumn("bill_category", bill_udf(col("bill_amount"))).show()
```

```python
#Exercise 73
def doc_fee(consultation_fee):
  if consultation_fee < 1000:
    return "Basic"
  elif consultation_fee >= 1000 and consultation_fee <= 1400:
    return "Standard"
  else:
    return "Premium"

doc_udf= udf(doc_fee)
doctors.withColumn("fee_category", doc_udf(col("consultation_fee"))).show()
```

```python
#Exercise 74
def app_priority(status):
  if status == "Completed":
    return "Normal"
  elif status == "Pending":
    return "High"
  else:
    return "Low"

app_udf = udf(app_priority)
appointments.withColumn("appointment_priority", app_udf(col("status"))).show()
```

```python
#Exercise 75
def city_zone(city):
  if city in ["Hyderabad", "Chennai", "Bangalore"]:
    return "South"
  else:
    return "Metro"

zone_udf= udf(city_zone)
doctors.withColumn("City_zone", zone_udf(col("city"))).show()
```
