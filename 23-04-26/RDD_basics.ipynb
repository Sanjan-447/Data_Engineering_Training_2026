```python
rdd = sc.parallelize([1,2,3,4,5])
mapped_rdd = rdd.map(lambda x: x*2)
filtered_rdd = mapped_rdd.filter(lambda x: x > 5)

print("Nothing executed yet")
```

In the above code, all the mentioned functions are **transformations** which create a new RDD but do not execute the functions.

```python
filtered_rdd.collect()
```

Lazy Evaluation: It only executes and displays the process after performing an **action **

```python
rdd = sc.parallelize([1,2,3,4,5])
squared_rdd= rdd.map(lambda x: x*x)
squared_rdd.collect()
```

Map : it maps each element with the given data, it takes only 1 element

```python
rdd = sc.parallelize(["Rahul", "Arjun", "Sneha"])
upper_rdd = rdd.map(lambda x: x.upper())
upper_rdd.collect()
rdd.getNumPartitions()
```

```python
rdd = sc.parallelize([15, 20, 25, 30, 35, 40])
even_rdd= rdd.filter(lambda x: x % 2 == 0)
even_rdd.collect()
```

Filter: in order to filter the data based on the condition

```python
rdd= sc.parallelize(["Apple", "Banana", "Mango", "Kiwi"])
long_rdd= rdd.filter(lambda x : len(x) > 5)
long_rdd.collect()
```

```python
numbers= sc.parallelize([10,20,30,20,40,10,60,50,])
distinct_rdd= numbers.distinct()
distinct_rdd.collect()
```

Distinct : removes duplicates and gives the unique elements

```python
rdd1= sc.parallelize([1,2,4])
rdd2= sc.parallelize([3,5,6])
union_rdd = rdd1.union(rdd2)

union_rdd.collect()
```

Union : it adds the 2 rdd into 1 single rdd

```python
rdd1= sc.parallelize([1,2,3,4])
rdd2= sc.parallelize([3,4,5,6])

rdd1.intersection(rdd2).collect()
```

Reduce: used to reducec the elements into a single number

```python
data = [
("IT", 70000),
("HR", 60000),
("IT", 75000),
("Finance", 80000),
("HR", 58000)
]

rdd = sc.parallelize (data)
rdd.collect()
```

```python
rdd.reduceByKey(lambda a,b : a+b).collect()
```

ReduceByKey: it reduces all the similar key's values into 1 value and displays it, so there will be only unique keys.

```python
grouped = rdd.groupByKey()
[(k, list(v)) for k,v in grouped.collect() ]
```

groupByKey() : to group the multiple values for a key into a list or other data structure in order to keep it separate.

```python
rdd= sc.parallelize([
    ("Rahul", 80),
    ("Arjun", 92),
    ("Sneha", 86)
])

rdd.mapValues(lambda x: x+5).collect()
```

mapValues: used to update or do manipulations on the values of the data structure

```python
rdd = sc.parallelize([10,20,50,30,90])
rdd.sortBy(lambda x: x).collect()
```

sortBy: used to sort the values in the rdd


```python
lines= sc.parallelize([
    "python SQL spark",
    "hadoop spark hadoop",
    "python hadoop"
])

word_counts=(
    lines.flatMap(lambda line: line.split(" "))
    .map(lambda word: (word,1))
    .reduceByKey(lambda a,b: a+b)
)

word_counts.collect()
```

To find count of each word using multiple functions

```python
emp= sc.parallelize([
    ("Rahul", 80000),
    ("Arjun", 92000),
    ("Sneha", 86000),
    ("Priya", 45000),
    ("Ramesh", 54000)
])

high_salary= emp.filter(lambda x: x[1] > 65000)
high_salary.collect()
```
