```python
from pyspark.sql import SparkSession

spark= SparkSession.builder.appName("Part_6_RDD").getOrCreate()
sc= spark.sparkContext
print("Spark ready")
```

```python
employees = [
("Rahul","IT",70000),
("Sneha","HR",60000),
("Arjun","IT",75000),
("Priya","Finance",80000),
("Karan","IT",50000),
("Amit","HR",58000),
("Meera","Finance",82000),
("Ravi","IT",72000),
("Neha","HR",61000),
("Vikram","Finance",90000)
]
emp_rdd = sc.parallelize(employees)
emp_rdd.collect()
```

```python
#Exercise 29
emp_rdd.map(lambda x: (x[1], x[2])).reduceByKey(lambda a,b :a+b).sortBy(lambda x: x[1], ascending=False).take(1)
```

```python
#Exercise 30
emp_rdd.map(lambda x: (x[1], (1, x[2]))) \
       .reduceByKey(lambda a, b: (a[0] + b[0], a[1] + b[1])) \
       .collect()
```
