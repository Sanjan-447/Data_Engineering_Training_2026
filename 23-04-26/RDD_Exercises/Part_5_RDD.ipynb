```python
from pyspark.sql import SparkSession

spark= SparkSession.builder.appName("Part_5_RDD").getOrCreate()
sc= spark.sparkContext
print("Spark ready")
```

```python
employees = [
("Rahul","IT",70000),
("Sneha","HR",60000),
("Arjun","IT",75000),
("Priya","Finance",80000),
("Karan","IT",50000),
("Amit","HR",58000),
("Meera","Finance",82000),
("Ravi","IT",72000),
("Neha","HR",61000),
("Vikram","Finance",90000)
]
emp_rdd = sc.parallelize(employees)
emp_rdd.collect()
```

```python
#Exercise 24
emp_count=emp_rdd.map(lambda x: (x[1], 1)).reduceByKey(lambda x,y: x+y).collect()
for k,v in emp_count:
  print(f"{k} -> {v}")
```

```python
#Exercise 25
emp_rdd.sortBy(lambda x: x[2], ascending=False).collect()
```

```python
#Exercise 26
emp_rdd.sortBy(lambda x: x[2], ascending=False).take(3)
```

```python
#Exercise 27
emp_rdd.map(lambda x: (x[1],x[0])).collect()
```

```python
#Exercise 28
grp= emp_rdd.map(lambda x: (x[1], x[0])).groupByKey().mapValues(list)
grp.collect()
```
