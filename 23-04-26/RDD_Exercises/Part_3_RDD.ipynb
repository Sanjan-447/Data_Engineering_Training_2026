```python
from pyspark.sql import SparkSession

spark= SparkSession.builder.appName("Part_3_RDD").getOrCreate()
sc= spark.sparkContext
print("Spark ready")
```

```python
sales = [
("Laptop",1,75000),
("Mouse",3,500),
("Keyboard",2,1500),
("Laptop",1,75000),
("Monitor",1,12000),
("Mouse",2,500),
("Keyboard",1,1500),
("Laptop",1,75000),
("Monitor",2,12000),
("Mouse",3,500),
("Laptop",1,75000),
("Keyboard",2,1500),
("Monitor",1,12000),
("Mouse",1,500),
("Laptop",1,75000)
]
sales_rdd = sc.parallelize(sales)
sales_rdd.collect()
```

```python
#Exercise 11
sales_rev= sales_rdd.map(lambda x: (x[0], x[1] * x[2])).reduceByKey(lambda x,y :x+y)
sales_rev.collect()
```

```python
#Exercise 12
sales_rdd.map(lambda x: x[1] * x[2]).reduce(lambda a,b : a+b)
```

```python
# Exercise 13
sales_rdd.map(lambda x: x[0]).distinct().collect()
```

```python
# Exercise 14
sales_rdd.map(lambda x: (x[0], x[1])).filter( lambda x: x[1] > 1).collect()
```

```python
# Exercise 15
sales_qty= sales_rdd.map(lambda x: (x[0], x[1])).reduceByKey(lambda x, y: x+y).collect()
for product, qty in sales_qty:
  print(f"{product} -> {qty}")
```

```python
# Exercise 16
total_rev= sales_rdd.map(lambda x: (x[0], x[1] * x[2])).reduceByKey(lambda a,b : a+b)

for product, rev in total_rev.collect():
  print(f"{product} -> {rev}")
```

```python
# Exercise 17
total_rev.sortBy(lambda x: x[1] , ascending = False).take(1)
```
