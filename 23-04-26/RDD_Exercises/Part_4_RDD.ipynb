```python
from pyspark.sql import SparkSession

spark= SparkSession.builder.appName("Part_3_RDD").getOrCreate()
sc= spark.sparkContext
print("Spark ready")
```

```python
employees = [
("Rahul","IT",70000),
("Sneha","HR",60000),
("Arjun","IT",75000),
("Priya","Finance",80000),
("Karan","IT",50000),
("Amit","HR",58000),
("Meera","Finance",82000),
("Ravi","IT",72000),
("Neha","HR",61000),
("Vikram","Finance",90000)
]
emp_rdd = sc.parallelize(employees)
emp_rdd.collect()
```

```python
#Exercise 18
emp_rdd.map(lambda x : x[0]).collect()
```

```python
#Exercise 19
emp_rdd.filter(lambda x: x[2] > 70000).collect()
```

```python
#Exercise 20
emp_rdd.map(lambda x: x[1]).distinct().collect()
```

```python
#Exercise 21
emp_sal=emp_rdd.map(lambda x: (x[1], x[2])).reduceByKey(lambda x,y :x+y).collect()
for dept, sal in emp_sal:
  print(f"{dept} -> {sal}")
```

```python
#Exercise 22
emp_dept= emp_rdd.map(lambda x: (x[1], (x[2], 1))).reduceByKey(lambda a,b : (a[0] + b[0], a[1] + b[1]))
dept_avg= emp_dept.map(lambda x: (x[0], x[1][0] / x[1][1]))
dept_avg.collect()
```
