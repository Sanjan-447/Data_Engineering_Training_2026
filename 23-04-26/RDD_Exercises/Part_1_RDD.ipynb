```python
from pyspark.sql import SparkSession

spark= SparkSession.builder.appName("Part_1_RDD").getOrCreate()
sc= spark.sparkContext

print("Spark Ready")
```

```python
logs = [
"Rahul login",
"Sneha login",
"Arjun login",
"Rahul purchase",
"Priya login",
"Sneha logout",
"Rahul login",
"Karan login",
"Arjun purchase",
"Sneha login",
"Rahul logout",
"Amit login",
"Priya purchase",
"Karan purchase",
"Sneha login",
"Rahul purchase",
"Meera login",
"Arjun login",
"Sneha purchase",
"Rahul login"
]
logs_rdd = sc.parallelize(logs)
logs_rdd.collect()
```

```python
#Exercise 1
logs_rdd.count()
```

```python
#Exercise 2
logs_rdd.take(5)
```

```python
#Exercise 3
logs_rdd.map(lambda words : words.split(" ")[0]).collect()
```

```python
#Exercise 4
logs_rdd.map(lambda words: words.split(" ")[1]).collect()
```

```python
#Exercise 5
logs_rdd.map(lambda word : word.split(" ")[0]).distinct().collect()
```

```python
#Exercise 6
logs_rdd.filter(lambda word : word.split(" ")[1] == "login").count()
```

```python
#Exercise 7
logs_rdd.filter(lambda word: word.split(" ")[1] == "purchase").collect()
```
